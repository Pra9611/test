```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="CGPA",y="Score",data = data,hue="Placement")
plt.show()
```

```python
x = data.iloc[:,:-1]
y = data["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=38)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha=2)
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100 , pr.score(x_test,y_test)*100
```

```python
for i in range(1,20):
    pr2 = Perceptron(alpha=1)
    pr2.fit(x_train,y_train)
    print(pr2.score(x_train,y_train)*100 , pr2.score(x_test,y_test)*100)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
```
