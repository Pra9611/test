```python
import tensorflow
```

```python
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
from keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
cnn = Sequential()
```

```python
cnn.add(Conv2D(32,(3,3),input_shape=(64,64,3),activation="relu"))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Conv2D(16,(3,3),activation="relu"))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Flatten())
```

```python
cnn.add(Dense(64,activation='relu'))
cnn.add(Dense(32,activation='relu'))
cnn.add(Dense(16,activation='relu'))
cnn.add(Dense(8,activation='relu'))
cnn.add(Dense(4,activation='relu'))
cnn.add(Dense(1,activation='sigmoid'))
```

```python
cnn.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(r"C:\Users\pramo\Downloads\cats",
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(r"C:\Users\pramo\Downloads\dog vs cat",
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')

cnn.fit_generator(train_generator,samples_per_epoch=200,nb_epoch=20,validation_data=test_generator)
```

```python
from keras.preprocessing import image
import numpy as np
```

```python
img = image.load_img(r"C:\Users\pramo\Downloads\dog.jpg",target_size=(64,64))
img = image.img_to_array(img)
```

```python
img = np.expand_dims(img,axis=0)
```

```python
p = cnn.predict(img)
```

```python
if p[0][0] < 0.5:
    print("DOG")
else:
    print("CAT")
```
